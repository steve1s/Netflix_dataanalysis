## Make a copy of the dataset

## Drop Null Values

```python
df = df.dropna()
df.shape
```

## Print first 10Values

```python
df.head(10)
```

## Converting Date and Time format

```python
# Strip spaces first
df["date_added"] = df["date_added"].str.strip()

# Convert to datetime safely
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

# Extract parts
df["day_added"] = df["date_added"].dt.day
df["month_added"] = df["date_added"].dt.month
df["year_added"] = df["date_added"].dt.year
```

```python
df.head(10)
```

# DATA Visualisation

```python
plt.figure(figsize=(10, 10))
sns.countplot(x="type", data=netflix)
plt.title("Type")
plt.show()
```

## Rating of shows and movies

```python

plt.figure(figsize=(13, 13))
ax = sns.countplot(x="rating", data=netflix, order=netflix["rating"].value_counts().index)

# Rotate labels for readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")

plt.title("Rating")
plt.show()
```

## Relation between Type and Rating

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating', hue='type', data=netflix)
plt.title('Relation between Types and Ratings')
plt.show()
```

## pie chart for the Type: Movie and TV shows

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

## Pie-chart for Rating

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
